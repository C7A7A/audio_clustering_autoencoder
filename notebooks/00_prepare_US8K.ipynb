Files were in directories: fold1-fold10. In each directory there were files with diferrent classes. This script copies files to their class directories. So intead of directories fold1-fold10 there are directories like dog_bark, car_horn, children_playing etc.

```python
import os
import pandas as pd
import shutil
```

```python
DATASET_DIR_NAME = 'US8K_original'
NEW_DATASET_DIR_NAME = 'US8K'
CSV_FILE_NAME = "UrbanSound8K.csv"
```

```python
csv_file = os.path.join('..\\datasets', DATASET_DIR_NAME, CSV_FILE_NAME)
dataset_path = os.path.join('..\\datasets', DATASET_DIR_NAME)
new_dataset_path = os.path.join('..\\datasets', NEW_DATASET_DIR_NAME)
```

```python
df = pd.read_csv(csv_file)
df
```

```python
print("Copying files to their class directories...")
for index, row in df.iterrows():
    class_name = row['class']
    file_name = row['slice_file_name']
    fold_name = f"fold{row['fold']}"
    
    source_file_path = os.path.join(dataset_path, fold_name, file_name)
    destination_dir = os.path.join(new_dataset_path, class_name)
    
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)
        
    destination_file_path = os.path.join(destination_dir, file_name)
    
    shutil.copy(source_file_path, destination_file_path)
    
print("Files have been copied to directories by class")
```
